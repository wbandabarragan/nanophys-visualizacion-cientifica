```python
import matplotlib.cbook as cbook
import matplotlib.pyplot as plt

import numpy as np
```

```python
image_file = cbook.get_sample_data('/Users/webb/Dropbox/Yachay_Tech/NANOPHYS/nanophys-visualizacion-cientifica/cotopaxi.jpg')

type(image_file)
```

```python
image = plt.imread(image_file)

type(image)

print(image.shape)
```

```python
# Figure enviroment

fig, ax = plt.subplots(figsize=(6,6))
ax.imshow(image)
plt.show()
```

```python
print(image.shape)
```

```python
imageR = image[:,:,0]
imageB = image[:,:,1]
imageG = image[:,:,2]
```

```python
print(imageR.shape)
print(imageB.shape)
print(imageG.shape)
```

```python
fig, ax = plt.subplots()
ax.imshow(imageR, cmap="Reds")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.imshow(imageB, cmap="Blues")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.imshow(imageG, cmap="Greens")
plt.show()
```

```python
print(imageG)

print(np.max(imageG), np.min(imageG))
```

### Convert to 1D:

```python
imageG_1D = np.reshape(imageG, (imageG.size, 1))
print(imageG.shape, imageG_1D.shape)

imageR_1D = np.reshape(imageR, (imageR.size, 1))

imageB_1D = np.reshape(imageB, (imageB.size, 1))
```

```python
plt.hist(imageG_1D, bins=50, histtype='step', label='Green', color = "g")
plt.hist(imageR_1D, bins=50, histtype='step', label='Red', color = "r")
plt.hist(imageB_1D, bins=50, histtype='step', label='Blue', color = "b")

plt.legend()
plt.show()
```
